```python

directory = r"/Users/kat/Desktop/NSF_HDR_Hackathon/FILES/Copernicus_ENA_Satelite Maps_Training_Data/"

# Initialize a list to store DataFrames
result_dataset = []

for filename in os.listdir(directory):
    if filename.endswith(".nc"):
        file_path = os.path.join(directory, filename)
        print(f"Processing file: {file_path}")

        # Open the dataset
        ds = xr.open_dataset(file_path)

        # Extract the 'sla' variable and convert to a DataFrame
        sla = ds["sla"]  # Sea Level Anomaly
        df = sla.to_dataframe().reset_index()  # Convert to DataFrame with dimensions as columns

        # Add the filename (or date extracted from filename) to the DataFrame
        date_str = filename.split(".")[0]
        df["source_file"] = date_str  # Add source identifier if useful

        # Append the DataFrame to the list
        result_dataset.append(df)

# Combine all DataFrames into one
combined_df = pd.concat(result_dataset, ignore_index=True)

# Display the combined DataFrame
print(combined_df.head())

# Optional: Save the combined DataFrame to a CSV file
output_path = "sla_data_per_date.csv"
combined_df.to_csv(output_path, index=False)
print(f"Combined dataset saved to {output_path}")
```

```python
import os
import xarray as xr

# Define the directory containing the .nc files
directory = r"/Users/kat/Desktop/NSF_HDR_Hackathon/FILES/Copernicus_ENA_Satelite Maps_Training_Data/"

# Initialize a list to store datasets
datasets = []

for filename in os.listdir(directory):
    if filename.endswith(".nc"):
        file_path = os.path.join(directory, filename)
        print(f"Processing file: {file_path}")

        # Open the dataset
        ds = xr.open_dataset(file_path)

        # Select only the 'sla' variable
        sla = ds["sla"]

        # Add the dataset to the list
        datasets.append(sla)

# Combine all datasets along the time dimension
combined_sla = xr.concat(datasets, dim="time")

# Display the combined dataset
print(combined_sla)

# Optional: Save the combined dataset to a new NetCDF file
output_path = "combined_sla_data.nc"
combined_sla.to_netcdf(output_path)
print(f"Combined dataset saved to {output_path}")
```
