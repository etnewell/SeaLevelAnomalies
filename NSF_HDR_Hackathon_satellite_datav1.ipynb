```python
import netCDF4
import os
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
```

```python
# Define the directory containing the .nc files
directory = r"/Users/kat/Desktop/NSF_HDR_Hackathon/FILES/Copernicus_ENA_Satelite Maps_Training_Data/"

# Initialize a list to store datasets
datasets = []
for filename in os.listdir(directory):
    if filename.endswith(".nc"):
        file_path = os.path.join(directory, filename)

        # Open the dataset
        ds = xr.open_dataset(file_path)

        # Select only the 'sla' variable
        sla = ds["sla"]

        # Add the dataset to the list
        datasets.append(sla)

# Combine all datasets along the time dimension
combined_sla = xr.concat(datasets, dim="time")

# output_path = "combined_sla_data.nc"
# combined_sla.to_dataframe(output_path)
```

```python
ds_3 = combined_sla.to_dataset()
ds_3 = combined_sla.to_dataframe()
ds_final = ds_3.reset_index()
```
